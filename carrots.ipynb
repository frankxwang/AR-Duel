```python
import numpy
import pandas
import glob
from keras.models import Sequential
from keras.layers import *
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from scipy import misc
import numpy as np
```

```python
seed = 7
numImg = glob.glob("data/training/People/*.jpg") + glob.glob("data/training/Not People/*.jpg")
testImg = glob.glob("data/validation/People/*.jpg") + glob.glob("data/validation/Not People/*.jpg")
numpy.random.seed(seed)
img_width = 100
img_height = 100
epochs = 50
```

```python
datasetX = []
for num in numImg:
    getImg = misc.imread(num)
    getImg = misc.imresize(getImg, (img_width, img_height))
    getImg = np.divide(getImg, 255)
    datasetX.append(getImg)
#print(datasetX)
datasetY = np.zeros(len(numImg))
for num in range(len(glob.glob("data/training/People/*.jpg"))):
    datasetY[num]= 1
#print(datasetY)
datasetX = np.array(datasetX)
datasetY = np.array(datasetY)
```

```python
testX = []
for num in testImg:
    getImg = misc.imread(num)
    getImg = misc.imresize(getImg, (img_width, img_height))
    getImg = np.divide(getImg, 255)
    testX.append(getImg)
#print(testX)
testY = np.zeros(len(testImg))
for num in range(len(glob.glob("data/validation/People/*.jpg"))):
    testY[num]= 1
#print(testY)
testX = np.array(testX)
testY = np.array(testY)
```

```python
model = Sequential()

model.add(Convolution2D(32,(3,3),input_shape=(img_width,img_height, 3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))


model.add(Convolution2D(32,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))


model.add(Convolution2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
model.fit(datasetX, datasetY, epochs=epochs, batch_size=64)
```

```python
test_img = misc.imread("data/validation/People/im1271.jpg")
print(model.evaluate(np.array([misc.imresize(test_img, (100, 100))]), 
                     np.array([1]), 
                     batch_size=1))
```

```python
print(model.evaluate(testX, 
                     testY, 
                     batch_size=1))
```

```python
model.save("test.h5")
```
